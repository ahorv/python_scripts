# Histogramm Analysis

```python
import imgutil
from matplotlib import pyplot as plt

%matplotlib inline

image_stack = imgutil.load_images()
       
plt, axes = plt.subplots(1, 4)
axes[0].imshow(image_stack[0])
axes[1].imshow(image_stack[1])
axes[2].imshow(image_stack[2])
axes[3].imshow(image_stack[3])    
```

## Show Histogram with opencv

```python

import imgutil
%matplotlib inline

image_stack = imgutil.load_images()
imgutil.cv_hist(image_stack[0],'ldr-Drago')
imgutil.cv_hist(image_stack[1],'ldr-Durand')
imgutil.cv_hist(image_stack[2],'ldr-Mantiuk')
imgutil.cv_hist(image_stack[3],'ldr-Reinhard')




```
